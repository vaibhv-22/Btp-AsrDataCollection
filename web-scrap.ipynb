```python
def title(url, data_type):
    req = requests.get(url)
    ind_soup = BeautifulSoup(req.content, 'html.parser')
    
    # ind_table_html = ind_soup.find('page-body resp-container indexes')
    ind_table_html = ind_soup.find('div', attrs={'class': 'page-body resp-container indexes'})
    # span with class = 'teaser'
    # ok = ind_table_html.find_all('a')
    ok = ind_table_html.find_all('a', href=True)
    # make ok only the text part of class a
    ok = [x.text for x in ok]
    return ok

def desc(url, data_type):
    req = requests.get(url)
    ind_soup = BeautifulSoup(req.content, 'html.parser')
    # all span with class = 'teaser'
    ind_table_html = ind_soup.find_all('span', attrs={'class': 'teaser'})
    # make ind_table_html only the text part of class span
    ind_table_html = [x.text for x in ind_table_html]
    return ind_table_html
```

```python
def get_districts_data(state_links, raw_path, data_type):
    dist_files_list = []

    for state in state_links:
        while True:
            try:
                state_url = f"https://mnregaweb4.nic.in/netnrega/state_html/{state['href']}"
                state_req = requests.get(state_url)
                time.sleep(1)
                state_soup = BeautifulSoup(state_req.content, 'html.parser')

                state_table_html = state_soup.find('center')
                # print(state_table_html)
                district_links = state_table_html.findAll('a', href=True)
                time.sleep(1)
                state_path = os.path.join(raw_path, f"{state.contents[0].title()}")
                if not os.path.exists(state_path):
                       os.mkdir(state_path)


                # print(district_links[0:len(district_links)-1])
                for district in district_links[0:len(district_links)-1]:
                    dist_url = f"https://mnregaweb4.nic.in/netnrega/state_html/{district['href']}"
                    dist_req = requests.get(dist_url)
                    time.sleep(1)
                    dist_soup = BeautifulSoup(dist_req.content, 'html.parser')

                    dist_table_html = dist_soup.find('center')
                    
                    block_links = dist_table_html.findAll('a', href=True)
                    district_path = os.path.join(state_path, f"{district.contents[0].title()}")
                    if not os.path.exists(district_path):
                        os.mkdir(district_path)

                    for block in block_links[0:len(block_links)-1]:
                        block_url = f"https://mnregaweb4.nic.in/netnrega/state_html/{block['href']}"    
                        block_req = requests.get(block_url)
                        time.sleep(1)
                        block_soup = BeautifulSoup(block_req.content, 'html.parser')

                        block_table_html = block_soup.find('center')
                        # print(block_table_html)
                        panchayat_links = block_table_html.findAll('a', href=True)
                        print(panchayat_links)

                        block_table_list = pd.read_html(str(block_table_html))
                        # print(block_table_list)
                        # block_path = os.path.join(raw_path, r'block_'+data_type)
                        # if not os.path.exists(block_path):
                        #     os.mkdir(block_path)
                        if block.contents[0].title() == 'Gm Pore/Trespone':
                            block_name1 = 'GM-PORE-TRESPONE'
                        else:
                            block_name1 = f"{block.contents[0].title()}"
                        block_path = os.path.join(district_path, block_name1)
                        if not os.path.exists(block_path):
                            os.mkdir(block_path)

                        if len(block_table_list) > 0:
                            block_df = block_table_list[0]
                            block_df.columns = block_df.iloc[1]
                            block_df = block_df.iloc[3:].copy()
                            block_df.insert(1, 'State',state.contents[0])
                            block_df.insert(2, 'district',district.contents[0])
                            block_df.insert(3, 'block',block.contents[0])
                            block_df.loc[block_df.index[-1], 'State']= np.nan
                            block_df.loc[block_df.index[-1], 'district']= np.nan
                            display(block_df)
                            filename = os.path.join(block_path, f"{block_name1}.csv")
                            block_df.to_csv(filename, index=False)






                    districts_table_list = pd.read_html(str(dist_table_html))
                    # print(districts_table_list)
                    # dist_path = os.path.join(raw_path, r'dist_'+data_type)
                    # if not os.path.exists(dist_path):
                    #     os.mkdir(dist_path)

                    if len(districts_table_list) > 0:
                        dists_df = districts_table_list[0]
                        dists_df.columns = dists_df.iloc[1]
                        dists_df = dists_df.iloc[3:].copy()
                        dists_df.insert(1, 'State',state.contents[0])
                        dists_df.insert(2, 'district',district.contents[0])
                        dists_df.loc[dists_df.index[-1], 'State']= np.nan
                        # display(dists_df)






                districts_table_list = pd.read_html(str(state_table_html))
                # print(districts_table_list)
                # dist_path = os.path.join(raw_path, r'dist_'+data_type)
                # if not os.path.exists(dist_path):
                #     os.mkdir(dist_path)

                if len(districts_table_list) > 0:
                    dists_df = districts_table_list[0]
                    dists_df.columns = dists_df.iloc[1]
                    dists_df = dists_df.iloc[3:].copy()
                    dists_df.insert(1, 'State',state.contents[0])
                    dists_df.loc[dists_df.index[-1], 'State']= np.nan
                    # display(dists_df)
                    # filename = os.path.join(dist_path, f"{state.contents[0].title()}_{data_type}.csv")
                    # dists_df.to_csv(filename, index=False)
                    # dist_files_list.append(filename)
                # del state_req, state_soup, state_table_html, districts_table_list
                break
            except Exception as e:
                print(e)

    return dist_files_list
```
